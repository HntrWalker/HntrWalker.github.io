```
from matplotlib import colors
plt.figure(figsize=(10,7))
plt.hist(target, color=['green'])
plt.xticks(rotation=90)
plt.show()

# All groupings have the same amount of data points
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(scaled_df, key, test_size=0.2, random_state=42)
```

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score

def bestK(X_train, y_train, X_test, y_test, min_k = 1, max_k = 25):
  best_K = 0
  best_score = 0.0
  for k in range(min_k, max_k+1, 2):
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train, y_train)
    preds = knn.predict(X_test)
    f1 = f1_score(y_test, preds, average = 'micro')

    if f1 > best_score:
      best_k = k
      best_score = f1

  print("Best K: {}".format(best_k))
  print("F1-Score: {}".format(best_score))
```

```
bestK(X_train, y_train, X_test, y_test, min_k = 1, max_k = 25)
```

```
from sklearn.metrics import classification_report

knn = KNeighborsClassifier(n_neighbors=17)
knn.fit(X_train, y_train)
preds = knn.predict(X_test)

def metrics(labels, preds):
  print("Precision Score: {}".format(precision_score(labels, preds, average='micro')))
  print("Recall Score: {}".format(recall_score(labels, preds, average='micro')))
  print("Accuracy Score: {}".format(accuracy_score(labels, preds)))
  print("F1 Score: {}".format(f1_score(labels, preds, average='micro')))

metrics(y_test, preds)
```

```
print(classification_report(y_test, preds))
```
